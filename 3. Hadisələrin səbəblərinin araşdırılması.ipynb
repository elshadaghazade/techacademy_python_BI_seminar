```python
import pandas as pd, matplotlib.pyplot as plt
```

```python
file1 = "Stats19-Data1979-2004/Accidents7904.csv"

accidents = pd.read_csv(file1,
sep=',',
header=0,
index_col=False,
parse_dates=['Date'],
dayfirst=True,
tupleize_cols=False,
error_bad_lines=True,
warn_bad_lines=True,skip_blank_lines=True,
nrows=1000000)
```

# Hava şəraiti

## Hava şəraiti sözlər ilə

```python
weather_conditions = {
    1: 'Açıqdır, güclü külək yoxdur',
    2: 'Yağışlıdır, güclü külək yoxdur',
    3: 'Qar yağır, güclü külək yoxdur',
    4: 'Açıqdır, güclü külək var',
    5: 'Yağışlıdır, güclü külək var',
    6: 'Qarlıdır, güclü külək var',
    7: 'Duman və ya sis',
    8: 'Digər',
    9: 'Bilinmir'
}
```

## Hava şəraitində qəzaların ümumi sayı

```python
result = accidents.groupby('Weather_Conditions', as_index=False).count()
result['weather_conditions_str'] = result.Weather_Conditions.apply(lambda c: weather_conditions[c])
result = result.loc[:, ['weather_conditions_str', 
                        'Accident_Index']]
result.columns = ['weather_conditions', 'count']
result.sort_values('count', ascending=False)
```

## Hava şəraiti və günün müxtəlif saatlarında qəzalar

```python
result = accidents.groupby(['Weather_Conditions', 'Time'], as_index=False).count()
result['weather_conditions_str'] = result.Weather_Conditions.apply(lambda c: weather_conditions[c])
result = result.sort_values('Accident_Index', ascending=False)
result = result.loc[:, ['weather_conditions_str', 'Time', 'Accident_Index']]
result.columns = ['weather_condition', 'time', 'count']
result
```

## Hava şəraiti və həftənin günlərində qəzalar

```python
result = accidents.groupby(['Weather_Conditions', 'Day_of_Week'], as_index=False).count()
result['weather_conditions_str'] = result.Weather_Conditions.apply(lambda c: weather_conditions[c])
result = result.sort_values('Accident_Index', ascending=False)
result = result.loc[:, ['weather_conditions_str', 'Day_of_Week', 'Accident_Index']]
result.columns = ['weather_condition', 'day_of_week', 'count']
result
```

## Hava şəraiti, həftənin günləri və günün saatlarında qəzalar

```python
result = accidents.groupby(['Weather_Conditions', 'Day_of_Week', 'Time'], as_index=False).count()
result['weather_conditions_str'] = result.Weather_Conditions.apply(lambda c: weather_conditions[c])
result = result.sort_values('Accident_Index', ascending=False)
result = result.loc[:, ['weather_conditions_str', 'Day_of_Week', 'Time', 'Accident_Index']]
result.columns = ['weather_condition', 'day_of_week', 'time', 'count']
result
```

## Həftənin günlərinə görə itkilər

```python
casualty_count = accidents.groupby('Day_of_Week', as_index=False).Number_of_Casualties.count()
casualty_sum = accidents.groupby('Day_of_Week', as_index=False).Number_of_Casualties.sum()
casualty_probability = casualty_sum / casualty_count

fig = plt.figure(figsize=(8,4))
ax1 = fig.add_subplot(121)
ax1.set_xlabel('Day of Week')
ax1.set_ylabel('Casualty Count')
ax1.set_title('Casualties by Day of Week')
casualty_count.plot(kind='bar')

ax2 = fig.add_subplot(122)
casualty_probability.plot(kind='bar')
ax2.set_xlabel('Day of Week')
ax2.set_ylabel('Probability of Casualty')
ax2.set_title('Probability of Casualties by Day of Week')
```
