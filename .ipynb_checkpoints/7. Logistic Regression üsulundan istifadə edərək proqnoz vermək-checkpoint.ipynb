# Statistikada logistic regression texnikası Hə, Yox cavabları almaq üçün istifadə edilir. 
# Misalçün Ödəniş etmək, etməmək?

```python
import pandas as pd, numpy as np, matplotlib.pyplot as plt

heart = pd.read_csv("Heart.csv",
    sep=',',
    header=0,
    index_col=0,
    parse_dates=True,
    tupleize_cols=False,
    error_bad_lines=False,
    warn_bad_lines=False,
    skip_blank_lines=True,
    low_memory=False
)

heart.head()
```

```python
heart.dtypes
```

```python
heart.shape
```

### Sinə ağrılarını rəqəmlərə çeviririk

```python
agrilar = pd.Series({
    'asymptomatic' : 1, 
    'nonanginal' : 2, 
    'nontypical' : 3,
    'typical': 4
})

heart['ChestPain'] = heart['ChestPain'].map(agrilar)
heart.head()
```

### Thal sütununu rəqəmlərə çeviririk

```python
t = pd.Series({'fixed' : 1,
'normal' : 2,
'reversable' : 3})
heart['Thal'] = heart['Thal'].map(t)
heart.head()
```

### AHD (atherosclerotic heart disease) sütununu rəqəmlərə çeviririk

```python
t = pd.Series({'No' : 0,
'Yes' : 1})
heart['AHD'] = heart['AHD'].map(t)
heart.head()
```

```python
heart.fillna(0, inplace=True)
heart.head()
```

### Modelimiz üçün iki matris yaradırıq.
### heart_data - göstəricilər
### heart_targets - AHD var və ya yoxdur

```python
heart_data = heart.iloc[:,0:13].values
heart_targets = heart['AHD'].values
```

```python
from sklearn import linear_model
from sklearn.model_selection import train_test_split, cross_val_score

# model yaradırıq
logClassifier = linear_model.LogisticRegression(C=1, random_state=111)
# Əlimizdə olan dataların 80% modelimizi təlim etdirmək üçün istifadə edirik, 20% isə test üçün
X_train, X_test, y_train, y_test = train_test_split(heart_data, heart_targets, test_size=0.20, random_state=111)
# modelin dəqiqliyini hesablayırıq
scores = cross_val_score(logClassifier, heart_data, heart_targets, cv=12)
print("Accuracy: %0.2f (+/- %0.2f)" % (scores.mean(), scores.std() * 2))
```

```python
logClassifier.fit(X_train, y_train)
```

### Test datanı modeldən keçirərək predict edirik

```python
predicted = logClassifier.predict(X_test)
print("Predicted:", predicted)
```

### Dəqiqliyi ölçürük

```python
from sklearn import metrics
metrics.accuracy_score(y_test, predicted)
```

```python
# yuxarı-sol, aşağı sağ diaqonal düz predictionları göstərir
metrics.confusion_matrix(y_test, predicted)

# True Positives - 26; False Positives - 5
# False Negatives - 7; True Negatives - 23

# beləliklə bizim model 31-dən 26 positiv, 30-dan 7 neqativ prediction çıxardı
```
