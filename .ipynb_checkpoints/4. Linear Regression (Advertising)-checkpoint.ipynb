```python
import pandas as pd, matplotlib.pyplot as plt
```

```python
file1 = "./Advertising.csv"

ads = pd.read_csv(file1,
sep=',',
header=0,
index_col=False,
dayfirst=True,
tupleize_cols=False,
error_bad_lines=True,
warn_bad_lines=True,skip_blank_lines=True)
```

```python
ads.head()
```

# Televiziya reklamları ilə satışın arasında olan əlaqənin qrafiki

```python
ads.plot(kind='scatter', x='TV', y='sales', figsize=(16, 8))
ads.plot(kind='scatter', x='radio', y='sales', figsize=(16, 8))
ads.plot(kind='scatter', x='newspaper', y='sales', figsize=(16, 8))
```

# Linear regression

Linear regression statistikada dəyişənlər arasında əlaqələrin qiymətləndirilməsini araşdıran statistik bir prosesdir

```python
from sklearn.linear_model import LinearRegression

lm = LinearRegression()

features = ['TV', 'radio', 'newspaper']
x = ads[features]
y = ads.sales

# fit the data to the model
lm.fit(x, y)

# intercept - bütün x-lar sıfır olarsa y-in orta cavabıdır (televizor, radio və qəzet reklamları dayanarsa gəlirin orta qiyməti)
print("intercept:", lm.intercept_)

# modelin ölçmə dəqiqliyi
print("coef:", lm.coef_)

# aggregate the feature names to create single object
fc = list(zip(features, lm.coef_))
print("Coefficients:", fc)

# Calculate R-squared value
lm.score(x, y)

# make new prediction
lm.predict([[75.60, 132.70, 34]])
```
