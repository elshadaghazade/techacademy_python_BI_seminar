```python
import pandas as pd, numpy as np
```

```python
file1 = "Stats19-Data1979-2004/Accidents7904.csv"

accidents = pd.read_csv(file1,
sep=',',
header=0,
index_col=False,
parse_dates=['Date'],
dayfirst=True,
tupleize_cols=False,
error_bad_lines=True,
warn_bad_lines=True,skip_blank_lines=True,
nrows=1000000)
```

```python
dtypes = pd.DataFrame(accidents.dtypes, columns=['Data Types'])
dtypes
```

```python
missing_data_counts = pd.DataFrame(accidents.isnull().sum(), columns=['Missing Values'])
missing_data_counts
```

```python
present_data_counts = pd.DataFrame(accidents.count(), columns=['Present Data'])
present_data_counts
```

```python
unique_value_counts = pd.DataFrame(accidents.nunique(), columns=['Unique Values'])
unique_value_counts
```

```python
minimum_values = pd.DataFrame(accidents.min(), columns=['Minimum Value'])
minimum_values
```

```python
maximum_values = pd.DataFrame(accidents.max(), columns=['Maximum Value'])
maximum_values
```

```python
dtypes.join(present_data_counts)\
.join(missing_data_counts)\
.join(unique_value_counts)\
.join(minimum_values)\
.join(maximum_values)
```
