```python
import pandas as pd, numpy as np, matplotlib.pyplot as plt

colleges = pd.read_csv('College.csv',
    sep=',',
    header=0,
    index_col=0,
    parse_dates=True,
    tupleize_cols=False,
    error_bad_lines=False,
    warn_bad_lines=True,
    skip_blank_lines=True,
    low_memory=False
)

colleges.head()
```

```python
colleges.dtypes
```

## Aşağıdakı qrafikdən dərhal görmək olur ki, kənarda qalanlar var (qrafikin yuxarı hissəsi)

```python
# müraciətlər  və qəbul edilmiş müraciətçilər
colleges.boxplot(column=['Apps', 'Accept'], return_type='axes', figsize=(12,6))
```

```python
# Kənarda qalanları daha yaxşı görmək üçün scatter plotdan istifadə edək
colleges.plot(kind='scatter', x='Apps', y='Accept', figsize=(12,6))
```

## İki kənarda qalan müraciətçilər tapıldı - Rutgers at New Brunswick və Purdue University at West Lafayette kollecləridir 

```python
fig, ax = plt.subplots()
colleges.plot(kind='scatter',
    x='Accept',
    y='Apps',
    figsize=(16, 6),
    ax=ax
)


for k, v in colleges.iterrows():
    ax.annotate(k,(v['Accept'],v['Apps']))
    

fig.canvas.draw()
```
